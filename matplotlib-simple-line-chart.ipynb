# color's value #
- an RGB or RGBA tuple of float values in [0, 1] (e.g., (0.1, 0.2, 0.5) or (0.1, 0.2, 0.5, 0.3))
- a hex RGB or RGBA string (e.g., '#0F0F0F' or '#0F0F0F0F')
- a string representation of a float value in [0, 1] inclusive for gray level (e.g., '0.5')
- one of {'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'}
- a X11/CSS4 color name
- a name from the xkcd color survey prefixed with 'xkcd:' (e.g., 'xkcd:sky blue')
- one of {'C0', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9'}
- one of {'tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan'} which are the Tableau Colors from the ‘T10’ categorical palette (which is the default color cycle).
- All string specifications of color are case-insensitive.

```python
import matplotlib.pyplot as plt

fig = plt.figure(frameon=True)

x = [1,2,3,4,5]
y = [2,3,4,5,6]
plt.plot(x,y,label="label")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("title")
plt.grid(True)
# plt.label("label")
plt.show()
fig.show()
```

# Style of Line #

| 参数 | 说明 |
|---:|:---|
| linewidth | 线条宽度 |
| color | 线条颜色 | 
|linestyle | 线条风格 | 
| label | 线条图释 配合 legend 使用 |

[线条风格参考](http://www.cnblogs.com/darkknightzh/p/6117528.html)

## setup color ##

有三种表示颜色的方式:
a:用全名  
b:16进制如：#FF00FF  
c：RGB或RGBA元组（1,0,1,1） 
d：灰度强度如：‘0.7’

## setup style ##
| 符号 | 说明 |
|---:|:----|
|'-' | solid line style |
|'--'| dashed line style |
|'-.'| dash-dot line style |
|':' | dotted line style |

```python
import matplotlib.pyplot as plt

x1 = [1,2,3,4,5]
y1 = [2,3,4,5,6]

x2 = [1,4,5,3,6]
y2 = [5,4,2,1,5]
plt.plot(x1, y1, linewidth = '1', color='red', linestyle='-')
plt.plot(x2, y2, linewidth = '1', color='green')


plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("title")
plt.grid(True)
plt.show()
```

#  设置 图释 #

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,3,4,5,6]

x2 = [1,4,5,3,6]
y2 = [5,4,2,1,5]

plt.plot(x1, y1, linewidth = '1', color='red', label='text1')
plt.plot(x2, y2, linewidth = '1', color='green', label='text2')
plt.legend(loc='upper left')


plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("title")
plt.grid(True)
plt.show()
```
